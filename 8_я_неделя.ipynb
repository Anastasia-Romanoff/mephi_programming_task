```python
import math
import matplotlib.pyplot as plt

a1 = 15.7
a2 = 17.8
a3 = 0.71
a4 = 23.7
a5 = 34

isotopes = {
    "U " : (92,238),
    "Th" : (90, 232),
    "Ra" : (88, 226),
    "Cf" : (98,252)
}

def calc_energy(Z,A,a1,a2, a3, a4,a5):
  if A % 2 == 1:
        a5 = 0
  elif Z % 2 == 0 and (A - Z) % 2 == 0:
        a5 = 34
  else:
        a5 = -34

  G = (a1 * A - a2 * A**(2/3) - a3 * (Z**2)/A**(1/3) - a4 * ((A - 2*Z)**2) / A + a5 * A**(-3/4))
  return(G)
energies = {}
for name, (Z, A) in isotopes.items():
    energies[name] = calc_energy(Z,A,a1,a2, a3, a4,a5)

print(f'{energies} МэВ')

z_values = list(energies.keys())
energie_values = list(energies.values())

plt.figure(figsize=(10, 6))
plt.plot(z_values, energie_values, color='blue')
plt.xlabel('Z')
plt.ylabel('Энергия связи, МэВ')
plt.title('Зависимость энергии связи от заряда Z')
plt.grid(axis='y')
plt.show()
```
